## Visualize this dataset using a scatter plot

I have created a JSON file to easly read and change the input data. It can be found in root of this repository as named "data.json".

```python
import json
import matplotlib.pyplot as plt

with open('data.json') as data_file:
    data = json.load(data_file)

plt.plot(data['x1'], data['x2'], 'ro')
plt.axis([0, 9, 0, 9])
plt.title('Virtualization of given dataset')
plt.show()
```

## Propose a generative model

Draw random samples from a normal (Gaussian) distribution.

The probability density for the Gaussian distribution is

![](https://latex.codecogs.com/svg.latex?P%28x%29%20%3D%20%5Cfrac%7B1%7D%7B%7B%5Csigma%20%5Csqrt%20%7B2%5Cpi%20%7D%20%7D%7De%5E%7B%7B%7B%20-%20%5Cleft%28%20%7Bx%20-%20%5Cmu%20%7D%20%5Cright%29%5E2%20%7D%20%5Cmathord%7B%5Cleft/%20%7B%5Cvphantom%20%7B%7B%20-%20%5Cleft%28%20%7Bx%20-%20%5Cmu%20%7D%20%5Cright%29%5E2%20%7D%20%7B2%5Csigma%20%5E2%20%7D%7D%7D%20%5Cright.%20%5Ckern-%5Cnulldelimiterspace%7D%20%7B2%5Csigma%20%5E2%20%7D%7D%7D)

where $\mu$ is the mean and $\sigma$ the standard deviation. The square of the standard deviation, $\sigma^2$, is called the variance.


## Implement your generative model and sample data from your model

```python
import numpy as np

x1, x2 = data['x1'], data['x2']
matrix = np.matrix((x1,x2)).transpose()
sigma = matrix.std()
mu = matrix.mean()
sample = sigma * np.random.randn(100, 2) + mu
```

## Visualize synthetic data to show that your model generates a 'similar dataset'

```python
plt.plot(sample[:,0], sample[:,1], 'ro')
plt.axis([0, 9, 0, 9])
plt.title('Virtualization of synthetic data')
plt.show()
```
